### process for remove outliers

```python
# Z-score : for calculate probability of a score occuring within normal distribution. 
from scipy.stats import zscore
import numpy as np 
z = np.abs(zscore(df))
z.shape
```

```python
# points lie beyond +/- 3 that will be considered an outliers. 
threshold = 3.1
print(np.where(z>3.1))

```

```python
# saving the data which is present in +/-3 in a new variable. 

df_new = df[(z<3.1).all(axis = 1)]
print(df.shape)
print(df_new.shape)
```

```python
# Dividing data in features, class: 

x = df.drop("class", axis = 1)
y = df["class"]
```

```python
# Converting Data to remove skewness : 
from sklearn.preprocessing import power_transform
x = power_transform(x, method = 'yeo-johnson')
x
```

### scaling :

```python
# scale the data
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
# use binary classification model bec we have only 2 var. 
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
from sklearn.metrics import accuracy_score
```

```python
# Splitting the DataFrame

for i in range (0,1000): 
    x_train,x_test,y_train, y_test = train_test_split(x,y,random_state=i, test_size = 0.20)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,3) == round(accuracy_score(y_test,pred_test)*100,3): 
        print("at random_state", i, "the model performs well")
        print('Random state:', i)
        print("Training accuracy:", accuracy_score(y_train,pred_train)*100)
        print("Testing accuracy:", accuracy_score(y_test,pred_test)*100)
```

```python
# we use random state for more accuracy
# (x,y,test_size = 20, random_state= 210)
```

## Classification report

```python
# we check unseen data [test data]

from sklearn.metrics import classification_report
print(classification_report(y_test,pred_test))
```

### Cross validation score for logistic regression :

```python
# check if model is overfetted or not

pred_lr = lr.predict(x_test)
from sklearn.model_selection import cross_val_score
Iss = accuracy_score(y_test, pred_lr)
for m in range(2,11):
    Isscore = cross_val_score(lr,x,y,cv=m)
    Isc = Isscore.mean()
    print("at CV:", m)
    print("CrossValidation score:", Isc*100)
    print("accuracy Score:", Iss*100)
    
```

### AUC ROC Curve:

```python
# Checkng performance parameter to check model performance 

from sklearn.metrics import roc_curve, auc
fpr,tpr, thresholds = roc_curve(pred_test, y_test)
roc_auc = auc(fpr,tpr)
print(roc_auc)
```

```python
plt.figure()
plt.plot(fpr,tpr,color= "red", label="ROC curve(area=%0.2f)" %roc_auc)
plt.plot([0,1],[0,1], color= "green",lw=10,linestyle = "--")
plt.xlabel("False positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic")
plt.legend(loc="lower right")
plt.show()
         
```
